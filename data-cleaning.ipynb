```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books_df = pd.read_csv(f"{path}/books.csv")
```

```python
books_df
```

```python
missing_data =  books_df.isnull()
missing_data.sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(books_df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Data Pattern")
plt.show()
```

```python
import missingno as msno

# Dendrogram (clusters columns by missingness correlation)
msno.dendrogram(books_df, figsize=(10, 6))
plt.show()
```

```python
import numpy as np
```

```python
books_df["missing_description"] = np.where(books_df["description"].isna(), 1, 0)
books_df["age"] = 2025 - books_df["published_year"]
```

```python
columns_of_interest = ["num_pages","missing_description","average_rating","age"]
corr_matrix =  books_df[columns_of_interest].corr(method="spearman")

plt.figure(figsize=(6, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar=True, square=True)
plt.title("Spearman Correlation Matrix")
plt.show()
```

```python
books_df[
    (books_df["missing_description"] == 1)|
    (books_df["age"]).isna()|
    (books_df["num_pages"]).isna()|
    (books_df["average_rating"]).isna()
]
```

```python
new_books_df = books_df[
    ~(books_df["missing_description"] == 1) &
    ~(books_df["age"]).isna()&
    ~(books_df["num_pages"]).isna()&
    ~(books_df["average_rating"]).isna()
]
```

```python
new_books_df
```

```python
new_books_df["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
new_books_df["words_in_description"] = books_df["description"].str.split().str.len()
```

```python
new_books_df
```

```python
# new_books_df.loc[new_books_df["words_in_description"].between(25,34), "description"]
clean_book_df = new_books_df[new_books_df["words_in_description"]>24]
clean_book_df
```

```python
clean_book_df["title_and_subtitle"] = (
    np.where(clean_book_df["subtitle"].isna(), clean_book_df["title"],clean_book_df[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)

clean_book_df
```

```python
clean_book_df["tagged_description"] = clean_book_df[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
clean_book_df
```

```python
(
    clean_book_df.drop(["subtitle","missing_description","age","words_in_description"], axis=1).to_csv("books_cleaned.csv", index=False)
)
```
